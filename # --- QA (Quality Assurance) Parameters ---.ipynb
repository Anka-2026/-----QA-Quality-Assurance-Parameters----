```python
# --- QA (Quality Assurance) Parameters ---
```

```python
# Temsilcinin kullanması gereken kelimeler ve puan değerleri
```

```python
zorunlu_adimlar={"merhaba": 20, "yardimci": 30, "tesekkur": 20, "hoscakal": 30}
```

```python
# Temsilcinin asla söylememesi gereken kelimeler
```

```python
yasakli_kelimeler =["bilmiyorum", "yapamam", "bekle"]
```

```python
# Analiz edilecek örnek temsilci konuşması
```

```python
temsilci_konusmasi=["merhaba", "size", "nasil", "yardimci", "olabilirim", "bilmiyorum", "tesekkur", "ederim"]
```

```python
# Başlangıç değerleri
```

```python
toplam_kalite_puani=0
```

```python
eksik_adimlar=([])
```

```python
print("--- ÇAĞRI KALİTE DENETİMİ BAŞLADI ---")
```

```python
# ---  DENETİM MOTORU (ANALİZ) ---
```

```python
#Zorunlu Adım Kontrolü
```

```python
for adim in zorunlu_adimlar:
    if adim in temsilci_konusmasi:
            puan=zorunlu_adimlar[adim]
            toplam_kalite_puani=toplam_kalite_puani+puan
    print("BAŞARILI:", adim, "adımı tamamlandı.")
else:
    eksik_adimlar.append(adim)
print("EKSİK:", adim, "adımı atlanmış!")
```

```python
#Yasaklı Kelime  Kontrolü
```

```python
for kelime in temsilci_konusmasi:
    if kelime in yasakli_kelimeler:
        toplam_kalite_puani=toplam_kalite_puani-50
        print("UYARI: Yasaklı kelime kullanıldı!")
```

```python
#Yönetici Raporu
```

```python
print("----------------------------")
print("TOPLAM KALİTE SKORU:", toplam_kalite_puani)
if toplam_kalite_puani >= 80:
    print("DENETİM SONUCU: MÜKEMMEL ÇAĞRI")
elif toplam_kalite_puani >= 50:
    print("DENETİM SONUCU: GELİŞTİRİLMELİ")
else:
    print("DENETİM SONUCU: BAŞARISIZ / EĞİTİM GEREKLİ")

print("Temsilciye Geri Bildirim (Unutulanlar):", eksik_adimlar)
print("----------------------------")
```
